```python
class PriorityQueue:
    def __init__(self):
        self.container = []

    def insert(self, number):
        self.container.append(number)

    def min(self):
        if self.container:
            return min(self.container)
        else:
            return None

    def removeMin(self):
        if self.container:
            self.container.remove(self.min())

    def isEmpty(self):
        return len(self.container) == 0

    def __len__(self):
        return len(self.container)

pq = PriorityQueue()

while True:
    print("1. Insert a number")
    print("2. Print the smallest number")
    print("3. Remove the smallest number")
    print("4. Check if the container is empty")
    print("5. Check the length of the container")
    print("6. Quit")
    q19 = input("Enter your choice: ")

    if q19 == '1':
        n = int(input("Enter a number to insert: "))
        pq.insert(n)
    elif q19 == '2':
        print(f"The smallest number is: {pq.min()}")
    elif q19 == '3':
        pq.removeMin()
        print("The smallest number has been removed.")
    elif q19 == '4':
        print(f"Is the container empty? {pq.isEmpty()}")
    elif q19 == '5':
        print(f"The length of the container is: {len(pq)}")
    elif q19 == '6':
        break
    else:
        print("Invalid choice. Please try again.")
```
